```python
import numpy as np
import pandas as pd

# Function to calculate the trimmed mean for RMSE columns
def trimmed_mean_rmse(data, column_name):
    rmse_values = data[column_name].dropna()
    if len(rmse_values) > 2:
        rmse_values = rmse_values.sort_values()
        trimmed_rmse = rmse_values[1:-1]  # Remove the smallest and largest values
        return trimmed_rmse.mean()
    else:
        return np.nan  # Return NaN if not enough data to trim

# Initialize a DataFrame to store the results
results_df = pd.DataFrame()

# Process each sheet and calculate trimmed mean for RMSE columns
for sheet_name in xls.sheet_names:
    data = pd.read_excel(xls, sheet_name)
    rmse_columns = [col for col in data.columns if 'RMSE' in col]
    sheet_results = {}
    for col in rmse_columns:
        trimmed_mean = trimmed_mean_rmse(data, col)
        sheet_results[col] = trimmed_mean
    results_df = results_df._append(pd.DataFrame(sheet_results, index=[sheet_name]))

# Save the results to an Excel file
results_df.to_excel('trimmed_mean_RMSE_results.xlsx')
```

```python
import numpy as np
import pandas as pd

# Function to calculate the trimmed mean for RMSE columns
def trimmed_mean_rmse(data, column_name):
    rmse_values = data[column_name].dropna()
    if len(rmse_values) > 4:
        rmse_values = rmse_values.sort_values()
        trimmed_rmse = rmse_values[2:-2]  # Remove the two smallest and two largest values
        return trimmed_rmse.mean()
    else:
        return np.nan  # Return NaN if not enough data to trim

# Initialize a DataFrame to store the results
results_df = pd.DataFrame()

# Process each sheet and calculate trimmed mean for RMSE columns
for sheet_name in xls.sheet_names:
    data = pd.read_excel(xls, sheet_name)
    rmse_columns = [col for col in data.columns if 'RMSE' in col]
    sheet_results = {}
    for col in rmse_columns:
        trimmed_mean = trimmed_mean_rmse(data, col)
        sheet_results[col] = trimmed_mean
    results_df = results_df._append(pd.DataFrame(sheet_results, index=[sheet_name]))

# Save the results to an Excel file
results_df.to_excel('trimmed_mean_RMSE_results2.xlsx')
```
